<span style="color:red">TODO: Fix the malformed "CÉDULA" values.</span>

```python
def change_ced(x):
  if re.match("\d{2}-\d{3}",x['CÉDULA']) is None:
    x['CÉDULA']=x['CÉDULA'][:2]+'-'+x['CÉDULA'][2]+x['CÉDULA'][4:]
  return x 
homicides_df=homicides_df.apply(change_ced, axis=1)
```

<span style="color:red">TODO: Make something similar to check and fix the "CORREO" column (PATTERN: 4 digits before the @, only .edu.co and .com domains are allowed)</span>

```python
def change_email(x):
  if re.match("[A-Za-z]*\d{4}@+[A-Za-z]*",x['CORREO']) is None:
    x['CORREO']=x['CORREO'].replace("@","0@")
  if  not (x['CORREO'].find('.edu.co')>-1 or x['CORREO'].find('.com')>-1):
    x['CORREO']=x['CORREO'][:x['CORREO'].find('.')+1] + 'com'
  return x

homicides_df=homicides_df.apply(change_email, axis=1)
```

<span style="color:red">TODO: Create a function to fix the digitation errors for "DIA" column.</span>

```python
dict_days=['Lunes','Martes','Miércoles','Jueves','Viernes','Sábado', 'Domingo']
def change_day(x):
  distancias_dict=[]
  for i in dict_days:
    distancias_dict.append(pylev.levenshtein(x['DIA'],i))
  x['DIA']=dict_days[np.argmin(distancias_dict)]
  return x
```

```python
homicides_df=homicides_df.apply(change_day,axis=1)
```
